## Python คือเครื่องคิดเลขดีๆนี่เอง

เราสามารถคำนวณเลขง่ายๆได้เลยด้วย Python

*   add
*   subtract
*   multiply
*   division
*   exponent
*   modulo

```
# add บวก
print(1 + 1)
print(3 + 8)
```

```
# subtract ลบ
print(19 - 9)
print(20 - 3)
```

```
# multiply คูณ
print(2 * 5)
print(3 * 3)
```

```
# division หาร
# เด๋วเราจะอธิบายความแตกต่างเรื่อง int vs. float ในวีดีโอต่อๆไป
print(5/2.0)
print(12/4.0)
```

```
# exponent ยกกำลัง
print(3 ** 2) # สามยกกำลังสอง
print(4 ** 3) # สี่ยกกำลังสาม
```

```
# modulo หรือเรียกสั้นๆว่า mod คือการหารตัวเลข
# ถ้าหารลงตัวจะได้ค่าศูนย์ ถ้าหารไม่ลงตัว Python จะ return เศษกลับมาให้เรา

print(10 % 3) # สิบหารสาม เหลือเศษ 1
print(5 % 3) # ห้าหารสาม เหลือเศษ 2
print(6 % 2) # หารลงตัว 0
```

## เราสามารถสร้างการคำนวณที่ซับซ้อนขึ้นได้ 
ใช้วงเล็บ (  ) เข้ามาช่วยก็ได้ถ้าเกิดการคำนวณเรายาวๆหน่อย

```
# ตัวอย่าง 1
answer = (100 * 2) + (100 ** 2)
print(answer)
```

```
# ตัวอย่าง 2
answer = 100 + 2 * 3 ** 3
print(answer)

print("\nอธิบาย:\nเครื่องหมาย ** จะเกิดก่อน * และ +")
print("3 ** 3 = 27 แล้วก็คูณสองได้ 54 แล้วค่อยบวก 100")
print("คำตอบเลยได้เท่ากับ 154")
```

## Summary
Python คือเครื่องคิดเลข เราสามารถ print() ผลลัพธ์ของการคำนวณต่างๆของเราได้ง่ายๆเลย
