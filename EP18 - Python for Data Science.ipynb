### import library ที่ใช้ในงาน data science

ก่อนที่เราจะใช้ function ที่อยู่ใน numpy pandas ได้ เราต้อง import ฟังชั่นพวกนั้นเข้ามาใน Python notebook ของเราก่อน

```
import package_name as short_name
```

````
# load data science package in the first cell (always)
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
````

````
# load example data (iris)
iris = datasets.load_iris()
````

### สร้าง dataframe ข้อมูลของเรา

````
# convert to DataFrame
df = pd.DataFrame(iris.data, columns=list(iris.feature_names))
````

````
# print first five rows
df.head()
````

````
# print last five rows
df.tail()
````

### ดูค่า summary statistics ง่ายๆด้วย .describe()

````
# summary statistics
df.describe()
````

### เราสามารถสร้าง chart ง่ายๆด้วย .plot()

````
df.plot(x="sepal length (cm)", y="sepal width (cm)", kind="scatter", title="My first scatter plot")
````

````
# correlation
np.corrcoef(df["sepal length (cm)"], df["sepal width (cm)"])
````
