### function 4

```

    """return total sum of all numbers in list_a"""
    return value

### expected output
sum_list([1,2,3,4,5])
15
```



````
## function4 - your code here
def sum_list(list_a):
  """return total sum of all numbers in list_a"""
  value = 0
  
  for num in list_a:
    value += num
  return value

## test function
print(sum_list([1,2,3,4,5]))
print(sum_list([1,2,3,4,5,10]))
print(sum_list([1,2,3,4,5,100]))
````

````
## function7 - your code here
def grade_calc(score):
    """return grade of this student
    score >= 90 'A'
    score >= 80 'B'
    score >= 70 'C'
    score >= 60 'D'
    score < 60 'Failed'
    """
    if score >= 90:
      return "A"
    elif score >= 80:
      return "B"
    elif score >= 70:
      return "C"
    elif score >= 60:
      return "D"
    else:
      return "Failed"

### test function
print(grade_calc(99))
print(grade_calc(72))
print(grade_calc(52))
````

```
        modify your code in function 7

### expected output

['B', 'A', 'Failed']
grade_calc_list([20, 68, 75])

```

````
## function8 - your code here
def grade_calc_list(list_of_score):
    """return grade of student in a list
        modify your code in function 7
    """
    result = []
    
    for score in list_of_score:
      if score >= 90:
        result.append('A')
      elif score >= 80:
        result.append('B')
      elif score >= 70:
        result.append('C')
      elif score >= 60:
        result.append('D')
      else:
        result.append('Failed')
    
    return result
        
## test function
print(grade_calc_list([80, 92, 50]))
print(grade_calc_list([20, 68, 75]))
print(grade_calc_list([20, 99, 25]))
````

````
## function10 - your code here
def special_func(text):
    """
    How to write this function
      1. lower-case the text
      2. split text into list
      3. upper-case the word 'PYTHON'
      4. return the result
    """

    text = text.lower() # step1
    text = text.split() # step2

    result = []
    
    for word in text:
        if word == 'python':
            upper_word = word.upper() # step3
            result.append(upper_word)
        else:
            result.append(word)

    return result # step4
  
## test function
print(special_func("I AM lEARning python PROGramming"))
print(special_func("python is very fun"))
print(special_func("python is really really fun"))
````
